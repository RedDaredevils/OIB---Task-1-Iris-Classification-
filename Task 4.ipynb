```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
```

Loading dataset


```python
data=pd.read_csv("C:\\Users\\saadi\\Desktop\\spam.csv",encoding="latin-1")
```

```python
data['v1'].replace(regex={"spam":0,"ham":1},inplace=True)
```

```python
data
```

X and Y values


```python
x=data['v2']
y=data['v1']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1)
```

```python
cv = CountVectorizer()
fit_x = cv.fit_transform(x_train)
```

```python
model = svm.SVC()
model.fit(fit_x,y_train)
```

```python
predict_x = cv.transform(x_test)
print("accuracy",model.score(predict_x,y_test)*100)
```
